```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fifa = pd.read_csv('data/fifa_data.csv')
fifa.head(50)
```

```python
ticks = np.arange(40, 105, 5)
bins = np.arange(40, 101, 2.5)
binqtty = [1000, 2000, 3000, 4000, 5000]

plt.style.use('seaborn-v0_8')
plt.hist(fifa['Overall'], bins=bins, edgecolor='w')

plt.xticks(ticks)
plt.grid(axis='x')

plt.xlabel('Skill Level')
plt.ylabel('Number of players')
plt.title('Distribution of player skills in FIFA 2018')

plt.savefig('images/Distribution of player skills in FIFA 2018.jpg')
plt.show()
```

```python
left_foot = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right_foot = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left_foot, right_foot], labels=['Left foot', 'Right foot'], autopct='%.2f%%')
plt.title('Foot Preference of FIFA 2018 Players')

plt.savefig('images/Foot Preference of FIFA 2018 Players.jpg')
plt.show()
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa['Weight']]

light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] > 125) & (fifa['Weight'] < 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] > 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa['Weight'] > 175) & (fifa['Weight'] < 200)].count()[0]
heavy = fifa.loc[fifa['Weight'] > 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (.5, .1, 0, 0, .25)
plt.pie(weights, labels=labels, autopct='%.2f%%', pctdistance=.8, explode=explode)
plt.title('Weight Distribution of FIFA 2018 Players (in lbs)')

plt.savefig('images/Weight Distribution of FIFA 2018 Players (in lbs).jpg')
plt.show()
```

```python
Barcelona_Overall = fifa[fifa.Club == 'FC Barcelona']['Overall']
Madrid_Overall = fifa[fifa.Club == 'Real Madrid']['Overall']
Juventus_Overall = fifa[fifa.Club == 'Juventus']['Overall']

plt.boxplot(
    [Barcelona_Overall, Madrid_Overall, Juventus_Overall],
    labels=['FC Barcelona', 'Real Madrid', 'Juventus']
)

plt.grid(axis='x')
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.savefig('images/Professional Soccer Team Comparison.jpg')
plt.show()
```
